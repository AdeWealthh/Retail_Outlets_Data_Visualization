```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in daily customer data
daily_customers_df = pd.read_csv('https://tinyurl.com/ChrisCoDV/001250122/OutletDailyCustomers.csv', parse_dates=['Date'], dayfirst=True)

# Read in outlet marketing data
outlet_marketing_df = pd.read_csv('https://tinyurl.com/ChrisCoDV/001250122/OutletMarketing.csv')

# Read in outlet overheads data
outlet_overheads_df = pd.read_csv('https://tinyurl.com/ChrisCoDV/001250122/OutletOverheads.csv')

# Read in outlet size data
outlet_size_df = pd.read_csv('https://tinyurl.com/ChrisCoDV/001250122/OutletSize.csv')

# Read in outlet staff data
outlet_staff_df = pd.read_csv('https://tinyurl.com/ChrisCoDV/001250122/OutletStaff.csv')
```
